# Decomposing Time Series 

# Additive Model
y(t)=Level+Trend+Seasonality+Noise

# Multiplicative Model
y(t)=Level* Trend * Seasonality * Noise

```python
from statsmodels.tsa.seasonal import seasonal_decompose 
```

```python
import pandas as pd
```

```python
df=pd.read_csv('airline.csv',header=0,parse_dates=[0])
```

```python
df
```

```python
df.index=df['Month']
```

# Seasonal decompose function
to use seasonal decompose function we need to use two parameters.

1st parameter : Provide the series ( NOT the date time) 

2nd parameter : Select the model ( additive model)

```python
result=seasonal_decompose(df['MilesMM'],model='additive')
```

```python
result.plot()
```

```python
result1=seasonal_decompose(df['MilesMM'],model='multiplicative')
```

```python
result1.plot()
```
